```python
def get_synonyms_antonyms(word):
    """
    Get the synonyms and antonyms for a given word using WordNet.

    Parameters:
    - word (str): The word for which synonyms and antonyms are to be found.

    Returns:
    - tuple: A tuple containing two sets - synonyms and antonyms of the word.
    """
    synonyms = set()
    antonyms = set()

    # Loop through all synsets for the word
    for syn in wordnet.synsets(word):
        # Loop through all lemmas of the synset
        for lemma in syn.lemmas():
          # Add lemma name to synonyms set
            synonyms.add(lemma.name())
            # Check if the lemma has any antonyms
            if lemma.antonyms():
              # Add the antonym to the antonyms set
                antonyms.add(lemma.antonyms()[0].name())

    return synonyms, antonyms

def main():
    """
    Main function to get synonyms and antonyms for an input word.
    """
    # Download the wordnet data, necessary for fetching synonyms and antonyms
    nltk.download('wordnet')

    # Get user input for the word
    word = input("Enter a word: ")

    # Fetch synonyms and antonyms for the input word
    synonyms, antonyms = get_synonyms_antonyms(word)


    print(f"Synonyms of '{word}': {', '.join(synonyms)}")
    print(f"Antonyms of '{word}': {', '.join(antonyms)}")

# Check to ensure the script is being run as the main module
if __name__ == "__main__":
    main()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from collections import Counter

# Download necessary datasets for tokenization and stopwords
nltk.download('stopwords')
nltk.download('punkt')

# Set up the list of stop words
stop_words = set(stopwords.words('english'))

def preprocess_tweet(tweet):
    """
    Preprocesses a tweet by removing URLs, mentions, and hashtags.

    Parameters:
    - tweet (str): The tweet to preprocess.

    Returns:
    - str: The preprocessed tweet.
    """
    # Remove URLs
    tweet = re.sub(r"http\S+|www\S+|https\S+", "", tweet, flags=re.MULTILINE)
    # Remove mentions (@username) and hashtags (#)
    tweet = re.sub(r"\@\w+|\#", "", tweet)

    return tweet.strip()

def remove_stop_words(tweet):
    """
    Removes stop words from a tweet.

    Parameters:
    - tweet (str): The tweet from which stop words should be removed.

    Returns:
    - str: The tweet without stop words.
    """
    # Tokenize the tweet into individual words
    words = word_tokenize(tweet)
    # Remove words that are in the list of stop words
    filtered_words = [word for word in words if word.lower() not in stop_words]
    # Join the filtered words back into a string
    filtered_tweet = ' '.join(filtered_words)

    return filtered_tweet

def count_word_frequency(text):
    """
    Counts the frequency of each word in a text.

    Parameters:
    - text (str): The text for which to count word frequencies.

    Returns:
    - Counter: A dictionary-like object with word frequencies.
    """
    # Tokenize the text into individual words
    words = word_tokenize(text)
    # Count the frequency of each word
    word_frequency = Counter(words)

    return word_frequency

# Sample tweets
tweets = [
    "Just had an amazing lunch with friends! #happy",
    "Check out this awesome website: www.example.com",
    "Can't believe how terrible the weather is today. #disappointed"
]

for tweet in tweets:
    # Preprocess each tweet by removing URLs, mentions, and hashtags
    preprocessed_tweet = preprocess_tweet(tweet)
    # Remove stop words from the preprocessed tweet
    filtered_tweet = remove_stop_words(preprocessed_tweet)
    # Count the frequency of each word in the filtered tweet
    word_frequency = count_word_frequency(filtered_tweet)


    print("Original tweet:", tweet)
    print("Filtered tweet:", filtered_tweet)
    print("Word Frequency:", word_frequency)
    print()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist

# A subset of a corpus, which is a short excerpt from a story.
subset_corpus = "The quick brown fox jumps over the lazy dog"

# Tokenize the corpus and convert each token to lowercase.
# This ensures uniformity in counting the frequency of each word, regardless of its case.
tokens = word_tokenize(subset_corpus.lower())

# Create unigrams, bigrams, and trigrams from the tokenized corpus.
# Unigrams are single words, bigrams are pairs of words, and trigrams are triples of words.
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

# Calculate the frequency distribution for each set of n-grams.
# This will give us a count of how often each unigram, bigram, or trigram appears in the corpus.
unigram_freq = FreqDist(unigrams)
bigram_freq = FreqDist(bigrams)
trigram_freq = FreqDist(trigrams)

# Print the most common unigrams, bigrams, and trigrams in the corpus, along with their frequencies.
# In this case, we're showing the top 10 for each category.
print("Most Common Unigrams:")
print(unigram_freq.most_common(10))

print("\nMost Common Bigrams:")
print(bigram_freq.most_common(10))

print("\nMost Common Trigrams:")
print(trigram_freq.most_common(10))
```

```python
#reviews in English.
reviews = [
    "This product exceeded my expectations It's of high quality and very durable",
    "I'm extremely disappointed with this purchase The item arrived damaged",
    "The customer service was excellent They promptly answered my questions and resolved my issue",
    "I love this product It's a great value for the price",
    "The shipping took longer than expected but the product itself is fantastic",
]

# Tokenize each review in the reviews list.
# word_tokenize is a function from the NLTK library that breaks text into individual words.
tokenized_reviews = [word_tokenize(review) for review in reviews]

# Combine all the tokenized reviews into a single string.
# This combined text will be used to generate the word cloud.
combined_text = " ".join([" ".join(tokens) for tokens in tokenized_reviews])

# Printing the tokenized words for each review.
for i, tokens in enumerate(tokenized_reviews):
    print(f"Tokens for Review {i + 1}:")
    print(tokens)
    print()

# Generate a word cloud based on the combined text.
# WordCloud is a visualization where the size of each word indicates its frequency or importance.
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_text)
plt.figure(figsize=(10, 5))
# Display the generated word cloud.
plt.imshow(wordcloud, interpolation='bilinear')
# Remove the axis values.
plt.axis('off')
plt.show()
```
