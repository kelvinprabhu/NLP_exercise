#Word Vectorization

##The word vectorization or embedding is the process of converting text into numerical vectors in a vector space.
##Each numerical value corresponding to the words can preserve some semantic information about the word and that may be lexical or compositional semantics.
##* Sparse or frequency-based vectors - The sparse vector representation captures less semantic information with exploiting considerable amount of memory.
##Eg. - Bag of Words (BoW) or One Hot Encoding, Term Frequency and Inverse Document Frequency (Tf-Idf) and Distributed representations are the some of the sparse representation methods.

#	Bag of Words or One Hot Encoding
#One hot encoding or Bag of Words (BoW) is a binary vector, which means only one index holds the non-zero value to represent the presence of word and the remaining indices will be zero.
#It is also termed as **count vectorization** and dummy variables.
#This method is very effective to represent categorical values.
#There are mainly two shortcomings associated with the one hot encoding.
#Initially, the curse of dimensionality problem, when the data with a higher dimension then the matrix taken up to huge amount of zeros other than only one index with ‘1’ to represent the presence of word.
#This leads to the issue of loss of memory.
#The second one is this method is hard to extract the semantics.

#Max_df stands for maximum document frequency.
#Itignore words which occur frequently.
#These words could be like the word ‘the’ that occur in every document and does not provide and valuable information to our text classification or any other machine learning model and can be safely ignored.
#Max_df looks at how many documents contain the word and if it exceeds the max_df threshold then it is eliminated from the sparse matrix.

```python
#The words "NLP, is, Computer" have been removed from the sparse matrix as they occur in more than 1 document.
from sklearn.feature_extraction.text import CountVectorizer

document = ["NLP is a field of Computer Science and Linguistics",
            "NLP is fun",
            "NLP makes humans computer interaction"]

# Create a Vectorizer Object
vectorizer = CountVectorizer(max_df=1)

vectorizer.fit(document)

# Printing the identified Unique words along with their indices
print("Vocabulary: ", vectorizer.vocabulary_)

# Encode the Document
vector = vectorizer.transform(document)

# Summarizing the Encoded Texts
print("Encoded Document is:")
print(vector.toarray())
```

#Activation Function: Activation functions introduce non-linearity to the neural network, allowing it to learn complex patterns in the data. Without activation functions, the neural network would essentially be a series of linear operations, making it limited in its expressive power. Popular activation functions include ReLU (Rectified Linear Activation), Sigmoid, and Tanh.

#Weight: In a neural network, a weight represents the strength of the connection between two neurons. When a neural network is trained, these weights are adjusted to minimize the difference between the predicted output and the actual output. These weights essentially determine the importance of the input signals.

#Bias: Bias is similar to the intercept added in a linear equation. It's an additional parameter in the neural network which allows you to shift the activation function either to the left or right. This additional degree of freedom helps the model to better fit the data.

#Optimizer: Optimizers are algorithms or methods used to change the attributes of your neural network such as weights and learning rate in order to reduce the losses. They work by iteratively adjusting the parameters of the neural network based on the gradients of the loss function with respect to the network parameters. Examples of optimizers include Stochastic Gradient Descent (SGD), Adam, and RMSProp.


#Regularization: Regularization is a technique used to prevent overfitting in machine learning models. It adds a penalty term to the loss function, discouraging the model from fitting the training data too closely and thus improving its ability to generalize to unseen data. Common regularization techniques include L1 regularization (Lasso), L2 regularization (Ridge), and Dropout.


#Dropout: Dropout is a regularization technique used in neural networks to prevent overfitting. During training, dropout randomly sets a fraction of the neurons in a layer to zero. This means that their outputs are ignored during that particular forward and backward pass. By doing this, dropout prevents the neural network from relying too much on specific neurons, forcing it to learn more robust features.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape and normalize input data
x_train = x_train.reshape(x_train.shape[0], 28, 28)
x_test = x_test.reshape(x_test.shape[0], 28, 28)
x_train = x_train / 255
x_test = x_test / 255
```

```python
# One-hot encode target values
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Create LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(28,28)))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(x_train, y_train, epochs=3, validation_data=(x_test, y_test))

# Evaluate on test set
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

bidirectional.png

```python
model = tf.keras.Sequential([
    encoder,
    tf.keras.layers.Embedding(
        input_dim=len(encoder.get_vocabulary()),
        output_dim=64,
        # Use masking to handle the variable sequence lengths
        mask_zero=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
print([layer.supports_masking for layer in model.layers])
```

```python
# predict on a sample text without padding.

sample_text = ('The movie was cool. The animation and the graphics '
               'were out of this world. I would recommend this movie.')
predictions = model.predict(np.array([sample_text]))
print(predictions[0])
```

```python
# predict on a sample text with padding

padding = "the " * 2000
predictions = model.predict(np.array([sample_text, padding]))
print(predictions[0])
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```

```python
history = model.fit(train_dataset, epochs=10,
                    validation_data=test_dataset,
                    validation_steps=30)
```

```python
test_loss, test_acc = model.evaluate(test_dataset)

print('Test Loss:', test_loss)
print('Test Accuracy:', test_acc)
```

layered_bidirectional.png

```python
model = tf.keras.Sequential([
    encoder,
    tf.keras.layers.Embedding(len(encoder.get_vocabulary()), 64, mask_zero=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64,  return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```

```python
history = model.fit(train_dataset, epochs=10,
                    validation_data=test_dataset,
                    validation_steps=30)
```

```python
test_loss, test_acc = model.evaluate(test_dataset)

print('Test Loss:', test_loss)
print('Test Accuracy:', test_acc)
```

```python
# predict on a sample text without padding.

sample_text = ('The movie was not good. The animation and the graphics '
               'were terrible. I would not recommend this movie.')
predictions = model.predict(np.array([sample_text]))
print(predictions)
```
